```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("df1.csv")
df.head()
df.info()
```

```python
df["x"].unique()
```

```python
sns.displot(df[df["x"]=='A']['y'], kde=True, rug=True)
plt.title("Age Distribution with KDE and rug plot")
plt.show()
```

```python
#Displot di y
sns.displot(df['y'], kde=True, rug=True, color='purple')
plt.title("Age Distribution with KDE and Rug Plot")
plt.show()
```

```python
# KDE Plot 
sns.kdeplot(data=df, x='y', fill=True, color='green')
plt.title("Survived Density Plot")
plt.show()
```

```python
dt=pd.read_csv("df2.csv")
dt.head()
df.info()
df.shape
```

```python
numeric_db = dt.select_dtypes(include = ['int64', 'float64'])

correlation = numeric_db.corr()['f'].sort_values(ascending = False)


plt.figure(figsize = (10,8))
top_features = correlation[1:]
sns.barplot(x = top_features.values, y = top_features.index)
plt.title('Top 10 Features Correlated with column f')
plt.xlabel('Correlation Coefficient')
plt.tight_layout()
plt.show()

correlation.apply(lambda x: f"{x:.3f} ({'Strong' if abs(x) > 0.005 else 'Moderate' if abs(x) > 0.003 else 'Weak'})").head(10)
```

```python
plt.title("Scatterplot F-C")
plt.xlabel("c")
plt.ylabel("f")

plt.scatter(dt['f'], dt['c'], color='b', marker='x')
```

```python
# Scatter Matrix (Pair Plot) for a subset of numeric columns
pd.plotting.scatter_matrix(dt[['f', 'c', 'g', 'a', "x"]], figsize=(10, 10), alpha=0.7)
plt.suptitle("Scatter Matrix")
plt.show()
```
