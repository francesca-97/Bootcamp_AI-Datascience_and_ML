```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Carica il dataframe
df1 = pd.read_csv("df1.csv")

# Statistiche descrittive di 'y'
descrizione_y = df1['y'].describe()
print("Statistiche descrittive di 'y':")
print(descrizione_y)

```

```python

# Visualizza la distribuzione con un istogramma e KDE
plt.figure(figsize=(12, 6))
sns.displot(df1['y'], kde=True, rug=True,  bins= 30, color='skyblue')
plt.title("Distribuzione della variabile 'y' con KDE e Istogramma")
plt.xlabel("y")
plt.ylabel("Frequenza")
plt.show()

```

```python

# Boxplot per identificare visivamente gli outlier
plt.figure(figsize=(6, 6))
sns.boxplot(x=df1['y'], color='lightcoral')
plt.title("Boxplot della variabile 'y'")
plt.xlabel("y")
plt.show()
```

```python
# Violin Plot
plt.figure(figsize=(6, 6))
sns.violinplot(x=df1['y'], color='lightgreen')
plt.title("Violin Plot della variabile 'y'")
plt.xlabel("y")
plt.show()

```

```python
# Strip Plot
plt.figure(figsize=(6, 6))
sns.stripplot(x=df1['y'], color='blue', alpha=0.5)
plt.title("Strip Plot della variabile 'y'")
plt.xlabel("y")
plt.show()

```

```python
# CDF (Cumulative Distribution Function)
plt.figure(figsize=(8, 6))
sns.ecdfplot(df1['y'], color='purple')
plt.title("Funzione di distribuzione cumulativa (CDF) di 'y'")
plt.xlabel("y")
plt.ylabel("Frequenza cumulativa")
plt.show()
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Carica il dataframe
df2 = pd.read_csv("df2.csv")

# Calcola la correlazione tra 'f' e tutte le colonne numeriche
correlation = df2.corr()
print("Correlazione di 'f' con le altre colonne:")
print(correlation['f'])

```

```python
# Creazione di una heatmap per visualizzare le correlazioni tra tutte le variabili
plt.figure(figsize=(12, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Heatmap delle Correlazioni")
plt.show()
```

```python


# Imposta il layout della figura
fig, axes = plt.subplots(3, 2, figsize=(18, 20))  # 3 righe, 2 colonne

# Heatmap della correlazione
sns.heatmap(df2.corr(), annot=True, cmap="coolwarm", fmt=".2f", ax=axes[0, 0])
axes[0, 0].set_title("Heatmap delle Correlazioni")

# Pair Plot (solo colonne principali)
sns.pairplot(df2, vars=['a', 'b', 'c', 'd', 'f'], diag_kind='kde', corner=True)
plt.suptitle("Pair Plot", y=1.02)  # Sovrascrive il titolo globale, viene separato

# Bar Plot delle correlazioni con 'f'
correlation_with_f = df2.corr()['f'].sort_values(ascending=False)
axes[0, 1].bar(correlation_with_f.index[:-1], correlation_with_f[:-1], color='teal')
axes[0, 1].set_title("Correlazione delle Colonne con 'f'")
axes[0, 1].set_ylabel("Valore di Correlazione")
axes[0, 1].set_xlabel("Colonne")
axes[0, 1].tick_params(axis='x', rotation=45)

# Scatter Plot (esempio 'c' vs 'f')
sns.scatterplot(data=df2, x='c', y='f', ax=axes[1, 0], color='blue', alpha=0.6)
axes[1, 0].set_title("Scatter Plot: 'c' vs 'f'")
axes[1, 0].set_xlabel("c")
axes[1, 0].set_ylabel("f")

# Joint Plot (esempio 'a' vs 'f')
sns.regplot(data=df2, x='a', y='f', ax=axes[1, 1], color='purple')
axes[1, 1].set_title("Joint Plot: 'a' vs 'f'")
axes[1, 1].set_xlabel("a")
axes[1, 1].set_ylabel("f")

# Rimuove l'ultimo slot vuoto nella griglia
axes[2, 0].axis('off')
axes[2, 1].axis('off')

# Regola il layout
plt.tight_layout()
plt.show()
```
