# 1) Caricamento del DataFrame e Analisi delle Correlazioni


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Caricamento del DataFrame e rimozione della colonna "Unnamed"
df = pd.read_csv('df3.csv').drop(columns=['Unnamed: 0'])

# Calcola le correlazioni 
correlation_matrix = df.corr()

# Visualizza la correlazione della colonna 'f'
print(correlation_matrix['f'])
```

```python
# 1. Matrice di Correlazione
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Matrice di Correlazione')
plt.show()
```

Osservando la matrice di correlazione, notiamo quanto segue:

La colonna "d" ha una correlazione molto alta con la colonna "f" (0.93), evidenziata in rosso.

Le colonne "b" e "y" hanno correlazioni negative significative con la colonna "f" (-0.37), mostrate in blu.

Le altre colonne hanno correlazioni molto basse o quasi nulle con la colonna "f", evidenziate in grigio.

Questo conferma che la colonna "d"  il principale contributore alla creazione della colonna "f".

```python

# Creazione dei grafici a dispersione 
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Scatter plot di 'f' vs 'd' (correlazione pi alta)
sns.scatterplot(x='d', y='f', data=df, ax=axes[0])
axes[0].set_title('Scatter Plot di d vs f')
axes[0].set_xlabel('d')
axes[0].set_ylabel('f')

# Scatter plot di 'f' vs 'b' (correlazione negativa significativa)
sns.scatterplot(x='b', y='f', data=df, ax=axes[1])
axes[1].set_title('Scatter Plot di b vs f')
axes[1].set_xlabel('b')
axes[1].set_ylabel('f')

# Mostra i grafici
plt.tight_layout()
plt.show()
```

# 2) Grafico della funzione =1/ nell'intervallo da -10 a 10.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)
y = 1 / x

# Creo il grafico
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = 1/x', color='blue')

# Miglioramenti
plt.title('Grafico di y = 1/x')
plt.xlabel('x')
plt.ylabel('y')
plt.axvline(x=0, color='red', linestyle='--', label='Asintoto x=0')  # Asintoto verticale
plt.axhline(y=0, color='green', linestyle='--', label='y = 0')  # Linea orizzontale per y = 0
plt.xlim(-10, 10)  # Limiti dell'asse x
plt.ylim(-10, 10)  # Limiti dell'asse y
plt.legend()
plt.grid(True) #Aggiunge la griglia
plt.show()
```

- First Heatmap  

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generazione dei dati per la heatmap
x_vals = np.linspace(-5, 5, 100)
y_vals = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_vals, y_vals)

Z = X * Y

# Plot della heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(Z, cmap='coolwarm',)
plt.title('Heatmap con X*Y')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

```

- Second Heatmap  

```python

# Generazione dei dati per la heatmap
x_vals = np.linspace(-5, 5, 100)
y_vals = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_vals, y_vals)

Z = X**2 + Y**2


# Plot della heatmap con colormap coolwarm
plt.figure(figsize=(8, 6))
sns.heatmap(Z, cmap='coolwarm', cbar=True)
plt.title('Heatmap con X**2 + Y**2')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
